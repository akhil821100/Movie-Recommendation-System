```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')
```

```python
movies_ratings = pd.merge(left=ratings,right=movies,on='movieId')
```

```python
movies_ratings = movies_ratings[['userId', 'movieId', 'title', 'genres', 'rating', 'timestamp']]
```

```python
movies_ratings_pivot = pd.pivot_table(index='userId',columns='title',values='rating',data=movies_ratings)
movies_ratings_pivot
```

```python
booleanDf = movies_ratings_pivot.notna()
booleanDf
```

```python
frequent_itemsets = apriori(booleanDf, min_support=0.11, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
rules
```

```python
rules = rules.sort_values(by=['confidence','lift'], ascending=[False,False])
rules
```

```python
def get_recommendations(movies_list,output_data):
    recommendations = []
    for index, row in rules.iterrows():
        ok = False
        for x in row['antecedents']:
            if x in movies_list:
                ok = True
                break

        if ok:
            for x in row['consequents']:
                recommendations.append(x)
    temp = recommendations
    recommendations = []
    for i in temp:
        if len(recommendations)==4:
            break
        if i not in recommendations:
            recommendations.append(i)
    output_data.append(recommendations)
    return output_data
```

```python
input_data=pd.read_csv('sample_test.tsv', sep='\t')
```

```python
input_movies = input_data['movies']
temp = input_movies
input_movies = []
for i in temp:
    i = i.split('\n')
    input_movies.append(i)
```

```python
output_data = []
for i in input_movies:
    output_data = get_recommendations(i,output_data)
    while(len(output_data[-1])<4):
        for j in range(len(movies['title'])):
            if (movies['title'][j] not in output_data[-1] and movies['title'][j] not in i):
                output_data[-1].append(movies['title'][j])
                break;
```

```python
import csv

header = ['movies', 'recommendations']
output_data_csv = []

for i in range(0,len(input_movies)):
    output_data_csv.append([input_movies[i],",".join(output_data[i])])

with open('output.csv', 'w', encoding='UTF8') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    for i in output_data_csv:
        writer.writerow(i)
```
